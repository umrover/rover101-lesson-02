# LCM Publisher

In this tutorial, we are going to make an LCM publisher. A publisher sends objects to *topics*, named points where another component (the *subscriber*) may receive them. [A sample subscriber is provided](./02b - LCM Subscriber.ipynb), and should be started first.

Firstly, we import the `lcm` package:

```python
import lcm
```

Now, we must make an LCM type. The file [Simple.lcm](./Simple.lcm) provides a very simple LCM message type.

We need to compile it into Python code first (**NOTE** this is a shell command, not a Python construct!):

```python
!lcm-gen --python Simple.lcm
```

Once we've done that, we can import the Python class `Simple` from the module `Simple`, generated by `lcm-gen`:

```python
from Simple import Simple
```

Let's publish a message once every second with random data:

```python
import random
import time

TIME_TO_SEND = 60
clock = 0

lc = lcm.LCM()

while clock < TIME_TO_SEND:
    msg = Simple()
    msg.elapsed_time_s = clock
    msg.latitude_deg = random.uniform(42, 43)
    msg.longitude_deg = random.uniform(-84, -82)
    lc.publish('/simple_topic', msg.encode())
    time.sleep(1.0)
    clock = clock + 1
```

In the subscriber, you will see it print the values each time they are received.
